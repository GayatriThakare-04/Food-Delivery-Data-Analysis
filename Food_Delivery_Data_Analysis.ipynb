## 4. Membership Distribution

```python
final_df["membership"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.show()
```

### Business Insight

- Gold members contribute the highest share of revenue.
- This suggests that loyalty programs are effective and should be further promoted to increase customer spending.

## Outlier Removal

```python
Q1 = final_df["total_amount"].quantile(0.25)
Q3 = final_df["total_amount"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

final_df = final_df[
    (final_df["total_amount"] >= lower) &
    (final_df["total_amount"] <= upper)
]
```
